```python
import requests
from bs4 import BeautifulSoup
import exchanges_rate as er


## WEB SCRAPİNG
 
r = requests.get('https://uzmanpara.milliyet.com.tr/doviz-kurlari/')
source = BeautifulSoup(r.content,"lxml")
a=source.find("div",attrs={"class":"realTimeBoxL"}).text
b=a.replace("ALIŞ (TL)",' ')
c=b.replace(",",".")
guncel_dolar=float(c)

## Yukarıda Güncel Doları Web Uzerinden Çektim.

q=source.find_all("div",attrs={"class":"realTimeBoxL"})[1].text
x=q.replace("ALIŞ (TL)",' ')
y=x.replace(",",".")
guncel_euro=float(y)

## Yukarıda Güncel Euroyu Web Uzerinden Çektim.



class Vergi_Hesap:
    def __init__(self,Fiyat,Hesaplama_Türü,Para_Birimi):
        self.Fiyat=Fiyat
        self.Hesaplama_Türü=Hesaplama_Türü
        self.Para_Birimi=Para_Birimi
    
    def Vergili_Fiyat_Hesaplama(self):
        if(self.Para_Birimi=="DOLAR" and self.Hesaplama_Türü=="ithalat_kayıtlı"):
            Vergisiz_Fiyat=(self.Fiyat*guncel_dolar)
            Uygulanan_Toplam_Vergi=Vergisiz_Fiyat*er.İTHALAT_VERGİ
            Toplam_Tutar=Vergisiz_Fiyat+Uygulanan_Toplam_Vergi
            Toplam_Tutar=round(Toplam_Tutar)
            message="Alacağınız Telefon İçin Toplam Fiyat : {0} TL dir."
            print(message.format(Toplam_Tutar))
        
        elif(self.Para_Birimi=="EURO" and self.Hesaplama_Türü=="ithalat_kayıtlı"):
            Vergisiz_Fiyat=(self.Fiyat*guncel_euro)
            Uygulanan_Toplam_Vergi=Vergisiz_Fiyat*er.İTHALAT_VERGİ
            Toplam_Tutar=Vergisiz_Fiyat+Uygulanan_Toplam_Vergi
            Toplam_Tutar=round(Toplam_Tutar)
            message="Alacağınız Telefon İçin Toplam Fiyat : {0} TL dir."
            print(message.format(Toplam_Tutar))
        
        elif(self.Para_Birimi=="DOLAR" and self.Hesaplama_Türü=="pasaport_kayıtlı"):
            Vergisiz_Fiyat=(self.Fiyat*guncel_dolar)
            Uygulanan_Toplam_Vergi=Vergisiz_Fiyat*er.PASAPORT_VERGİ
            Toplam_Tutar=Vergisiz_Fiyat+Uygulanan_Toplam_Vergi
            Toplam_Tutar=round(Toplam_Tutar)
            message="Alacağınız Telefon İçin Toplam Fiyat : {0} TL dir."
            print(message.format(Toplam_Tutar))
        
        elif(self.Para_Birimi=="EURO" and self.Hesaplama_Türü=="pasaport_kayıtlı"):
            Vergisiz_Fiyat=(self.Fiyat*guncel_euro)
            Uygulanan_Toplam_Vergi=Vergisiz_Fiyat*er.PASAPORT_VERGİ
            Toplam_Tutar=Vergisiz_Fiyat+Uygulanan_Toplam_Vergi
            Toplam_Tutar=round(Toplam_Tutar)
            message="Alacağınız Telefon İçin Toplam Fiyat : {0} TL dir."
            print(message.format(Toplam_Tutar))
        else:
            print("Hesaplama Başarısız !")

            

        
        
        
Telefon_Fiyatı=input("Telefon Fiyatını Girebilir misiniz ?\n\n")
Telefonun_Kayıt_Turu=input("Telefonun Kayıt Turu Nedir ?\n\n1-ithalat_kayıtlı\n2-pasaport_kayıtlı\n")
Para_Kur=input("Hangi Kurdan Hesaplayacaksınız ?\n\n1-DOLAR\n2-EURO\n")
x=Vergi_Hesap(float(Telefon_Fiyatı),Telefonun_Kayıt_Turu,Para_Kur)
x.Vergili_Fiyat_Hesaplama()




    
```
